```python
loan=int(input())
credit_score=int(input())
debt=int(input())
if loan>=500000 and credit_score>=700 and debt<0.4*loan:
  print("Eligible")
elif loan>= 300000 and credit_score >= 600:
  print("Review")
else:
  print("Reject")
```

```python
amount =int(input())

if amount < 1000:
    discount = 0
elif amount < 5000:
    discount = 0.05
else:
    discount = 0.10

final_amount = amount - (amount * discount)

print(final_amount)
```

```python
sentence ="learn ai and ml fundamentals".split()
abbreviations = ["AI", "ML"]
result = []

for word in sentence:
    if word.upper() in abbreviations:
        result.append(word.upper())
    else:
        result.append(word.capitalize())

print(" ".join(result))
```

```python
text = "abc123xyz"
result = ""

for ch in text:
    if ch.isdigit():
        result += "#"
    else:
        result += ch

print(result)
```

```python
numbers = [1,2,3,4]
result = []

for n in numbers:
    if n % 2 == 0:
        result.append(n**2)
    else:
        result.append(n**3)

print(result)
```

```python
password = "admin123"
attempts = 3

while attempts > 0:
    pwd = input("Enter password: ")
    if pwd == password:
        print("Access Granted")
        break
    attempts -= 1
    print("Wrong password. Attempts left:", attempts)

if attempts == 0:
    print("Access Denied")
```

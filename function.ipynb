```python
def twice(func, x):
    return func(func(x))
def square(n):
    return n * n
print(twice(square, 2))
```

```python
def multiplier(k):
    def multiply(x):
        return x * k
    return multiply
a= multiplier(2)
print(a(5))
```

```python
def counter():
    count = 0
    def next():
        nonlocal count
        count += 1
        return count
    return next
c = counter()
print(c())
print(c())
print(c())
```

```python
def accumulator(start=0):
    total = start
    def add(value):
        nonlocal total
        total += value
        return total
    return add
acc =accumulator(10)
print(acc(5))
print(acc(10))
```

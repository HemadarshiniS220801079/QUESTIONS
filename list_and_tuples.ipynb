```python
data = [1, 2, 3, 2, 4, 1, 5]
result = []
for i in range(len(data)):
    if data[i] not in data[i+1:]:
        result.append(data[i])
print(result)
```

```python
data = ['a', 'b', 'c', 'd', 'e']
result = []
for i in range(0, len(data)-1, 2):
    result.append((data[i], data[i+1]))
print(result)
```

```python
n = [10, 20, 20, 5, 15]
unique = []
for x in n:
    if x not in unique:
        unique.append(x)
if len(unique) < 2:
    print(-1)
else:
    largest = max(unique)
    unique.remove(largest)
    second_largest = max(unique)
    print(n.index(second_largest))
```

```python
a = [1,2,3]
b = ['a','b','c','d']
result = []
length = min(len(a), len(b))
for i in range(length):
    result.append(a[i])
    result.append(b[i])
print(result)
```

```python
data = [1,2,3,4,5]
a = 2
n = len(data)
for _ in range(a):
    last = data[-1]
    for i in range(n-1, 0, -1):
        data[i] = data[i-1]
    data[0] = last
print(data)
```

```python
data = [1,2,3,4,5,6]
result = []
for x in data:
    if x % 2 == 0:
        result.append(x*x)
print(result)
```

```python
data = [5, 3, 2, -4, 6]
total = 0
for x in data:
    if x < 0:
        break
    total += x
print(total)
```

```python
correct = "admin"
attempts = 3
while attempts > 0:
    a= input("Enter password: ")
    if a == correct:
        print("Unlocked")
        break
    attempts -= 1
if attempts == 0:
    print("Locked")
```

```python
data = [3, 5, -2, 4, 6, -1, 2]
result = []
current_max = 0
for x in data:
    if x <= 0:
        current_max = 0
    else:
        if x > current_max:
            current_max = x
    result.append(current_max)
print(result)
```
